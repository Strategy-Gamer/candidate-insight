```python
import spacy
```

```python
nlp = spacy.load("en_core_web_lg")

# NLP breaking into tokens
doc = nlp("Adison White. I support reproductive rights. I support the right to life. I am pro-choice. I want to protect the life of unborn children. Brenna hates children.")

for token in doc:
    print(token.text)
```

```python
for sent in doc.sents:
    print(f"Sentence: {sent.text}")
```

```python
issue_stance_phrases = {
    "Abortion": {
        "support": [
            "pro-choice",
            "defends reproductive rights",
            "supports the right to choose",
            "fighting for freedom to choose"
        ],
        "oppose": [
            "anti-abortion",
            "opposes reproductive rights",
            "wants to ban abortion",
            "against the right to choose"
        ]
    },
    "Healthcare": {
        "support": [
            "supports medicare for all",
            "wants universal healthcare",
            "healthcare is a human right",
            "expanding access to care"
        ],
        "oppose": [
            "against government-run healthcare",
            "wants to repeal the ACA",
            "opposes medicare for all",
            "cutting healthcare spending"
        ]
    },
    # Add more issues here...
}

def detect_issue_and_stance(sentence, issue_stance_phrases, threshold=0.75):
    doc_sent = nlp(sentence)
    best_match = None
    best_score = 0

    for issue, stance_data in issue_stance_phrases.items():
        for stance, phrases in stance_data.items():
            for phrase in phrases:
                doc_phrase = nlp(phrase)
                score = doc_sent.similarity(doc_phrase)
                if score > best_score and score >= threshold:
                    best_match = {
                        "issue": issue,
                        "stance": stance,
                        "score": score,
                        "matched_phrase": phrase
                    }
                    best_score = score

    return best_match

text = "Fighting for the Freedom to Choose"
match = detect_issue_and_stance(text, issue_stance_phrases)

if match:
    print(f"Issue: {match['issue']}")
    print(f"Stance: {match['stance']}")
    print(f"Matched Phrase: {match['matched_phrase']}")
    print(f"Score: {match['score']:.3f}")
else:
    print("No match found")
```

```python
def precompute_phrase_vectors(issue_dict, nlp):
    phrase_vectors = []
    for issue, stance_map in issue_dict.items():
        for stance, phrases in stance_map.items():
            for phrase in phrases:
                vector = nlp(phrase).vector
                phrase_vectors.append({
                    "issue": issue,
                    "stance": stance,
                    "phrase": phrase,
                    "vector": vector
                })
    return phrase_vectors

def precompute_issue_vectors(trigger_dict, nlp):
    issue_vectors = []
    for issue, phrases in trigger_dict.items():
        for phrase in phrases:
            vector = nlp(phrase).vector
            issue_vectors.append({
                "issue": issue,
                "phrase": phrase,
                "vector": vector
            })
    return issue_vectors

import numpy as np
from concurrent.futures import ThreadPoolExecutor

def detect_issue(text, trigger_phrases_dict, cached_vectors, nlp, similarity_threshold=0.87, min_gap=0.03):
    exact = detect_issue_exact_with_conflict(text, trigger_phrases_dict)

    if exact is None:
        # No exact match → semantic fallback
        return detect_issue_semantic(text, cached_vectors, nlp, similarity_threshold, min_gap)

    elif "conflict" in exact:
        # Multiple exact matches → discard or log
        print(f"Ambiguous exact match: {exact['conflict']}")
        return None

    else:
        return exact  # single, clean exact match
    

def detect_issue_exact_with_conflict(text, trigger_phrases_dict):
    text_lower = text.lower()
    matched_issues = []

    for issue, phrases in trigger_phrases_dict.items():
        for phrase in phrases:
            if phrase.lower() in text_lower:
                matched_issues.append(issue)
                break  # Stop checking more phrases for this issue

    if len(matched_issues) == 1:
        return {"issue": matched_issues[0], "confidence": 1.0, "method": "exact"}
    elif len(matched_issues) > 1:
        # Ambiguous: multiple exact matches
        return {"conflict": matched_issues}
    else:
        return None
    
def detect_issue_semantic(text, cached_issue_vectors, nlp, threshold=0.87, min_gap=0.03):
    doc_vector = nlp(text).vector

    def cosine_sim(vec1, vec2):
        return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2) + 1e-8)

    results = []
    for entry in cached_issue_vectors:
        score = cosine_sim(doc_vector, entry["vector"])
        if score >= threshold:
            results.append((score, entry["issue"]))

    if not results:
        return None

    results.sort(reverse=True)

    # Ambiguity check
    if len(results) > 1 and results[0][0] - results[1][0] < min_gap:
        return None  # Discard ambiguous match

    return {
        "issue": results[0][1],
        "confidence": results[0][0],
        "method": "semantic"
    }
    
def detect_issue_and_stance_fast(text, phrase_vectors, nlp, threshold=0.85):
    doc_vector = nlp(text).vector
    best_match = None
    best_score = -1

    def compare_vector(pv):
        score = np.dot(doc_vector, pv["vector"]) / (
            np.linalg.norm(doc_vector) * np.linalg.norm(pv["vector"]) + 1e-8
        )
        return (score, pv)

    with ThreadPoolExecutor() as executor:
        results = list(executor.map(compare_vector, phrase_vectors))

    for score, pv in results:
        if score > best_score:
            best_match = {
                "issue": pv["issue"],
                "stance": pv["stance"],
                "matched_phrase": pv["phrase"],
                "score": score
            }
            best_score = score

    if best_score >= threshold:
        return best_match
    else:
        return None
```

```python
from concurrent.futures import ThreadPoolExecutor

def detect_likely_issue(text, issue_dict, nlp, threshold=0.85):
    doc = nlp(text)
    best_issue = None
    best_score = 0

    for issue, stance_map in issue_dict.items():
        all_phrases = list(set(p for phrases in stance_map.values() for p in phrases))
        for phrase in all_phrases:
            score = doc.similarity(nlp(phrase))
            if score > best_score:
                best_issue = issue
                best_score = score

    if best_score >= threshold:
        return best_issue, best_score
    else:
        return None, None

def detect_likely_issue_parallel(text, issue_dict, nlp, threshold=0.85):
    doc = nlp(text)
    best_issue = None
    best_score = 0

    def score_phrase(issue, phrase):
        phrase_doc = nlp(phrase)
        return issue, doc.similarity(phrase_doc)

    with ThreadPoolExecutor() as executor:
        futures = []
        for issue, stance_map in issue_dict.items():
            phrases = set(p for plist in stance_map.values() for p in plist)
            for phrase in phrases:
                futures.append(executor.submit(score_phrase, issue, phrase))

        for future in futures:
            issue, score = future.result()
            if score > best_score:
                best_score = score
                best_issue = issue

    if best_score >= threshold:
        return best_issue, best_score
    else:
        return None, None
    
def detect_stance_for_issue(text, issue, issue_dict, nlp, threshold=0.75):
    doc = nlp(text)
    best_stance = None
    best_score = 0

    for stance, phrases in issue_dict[issue].items():
        for phrase in phrases:
            score = doc.similarity(nlp(phrase))
            if score > best_score and score >= threshold:
                best_stance = stance
                best_score = score

    return best_stance


# Sample mini issue set for test
for sent in doc.sents:
    sentence = sent.text    
    issue, score = detect_likely_issue_parallel(sentence, issue_stance_phrases, nlp)

    print(f"Sentence: {sentence}")
    if issue:
        stance = detect_stance_for_issue(sentence, issue, issue_stance_phrases, nlp)
        print(f"Issue: {issue} | Stance: {stance} | Confidence: {score:.2f}")
    else:
        print("No confident issue match found.")
```

```python
def detect_issue_and_stance_parallel(text, issue_dict, nlp, threshold=0.85):
    doc = nlp(text)
    best_match = None
    best_score = 0

    def compare_phrase(issue, stance, phrase):
        phrase_doc = nlp(phrase)
        return {
            "issue": issue,
            "stance": stance,
            "matched_phrase": phrase,
            "score": doc.similarity(phrase_doc)
        }

    with ThreadPoolExecutor() as executor:
        futures = []
        for issue, stance_map in issue_dict.items():
            for stance, phrases in stance_map.items():
                for phrase in phrases:
                    futures.append(executor.submit(compare_phrase, issue, stance, phrase))

        for future in futures:
            result = future.result()
            if result["score"] > best_score:
                best_match = result
                best_score = result["score"]

    if best_match and best_score >= threshold:
        return best_match
    else:
        return None
    
for sent in doc.sents:
    sentence = sent.text    
    result = detect_issue_and_stance_parallel(sentence, issue_stance_phrases, nlp)

    print(f"Sentence: {sentence}")
    if result:
        print(f"Issue: {result['issue']} | Stance: {result['stance']} | Score: {result['score']:.2f}")
    else:
        print("No confident match.")
```
